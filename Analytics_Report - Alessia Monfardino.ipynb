## Which country shows the best short-term opportunity for expansion? 
In order to answer this question, I developed the following questions:

### Which countries have the highest recorded online sales?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
Supply_Chain = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='unicode_escape')

# Keep only the necessary columns
Supply_Chain = Supply_Chain[['Order City', 'Order State', 'Order Country', 'order date (DateOrders)', 'Order Profit Per Order', 'Order Item Total']]

# Remove rows where Order City, Order State, or Order Country are blank
Supply_Chain = Supply_Chain.dropna(subset=['Order City', 'Order State', 'Order Country'])

# Format the order date to month/day/year format and remove the time
Supply_Chain['order date (DateOrders)'] = pd.to_datetime(Supply_Chain['order date (DateOrders)']).dt.strftime('%m/%d/%Y')

# Replace blank cells in Order Profit Per Order with the statistical median of the column
profit_median = Supply_Chain['Order Profit Per Order'].median()
Supply_Chain['Order Profit Per Order'] = Supply_Chain['Order Profit Per Order'].replace(np.nan, profit_median).round(2)

# Replace blank cells in Order Item Total with the statistical median of the column
item_total_median = Supply_Chain['Order Item Total'].median()
Supply_Chain['Order Item Total'] = Supply_Chain['Order Item Total'].replace(np.nan, item_total_median).round(2)

# Which countries have the highest order volume? 

# We can answer this question by grouping the data by country and counting the number of orders in each country:
order_volume = Supply_Chain.groupby('Order Country')['Order Country'].count().sort_values(ascending=False)[:10]

colors = ['#FEE4E4', '#FECDD3', '#FDC7DE', '#E2D2F0', '#D3E0F5', '#C9DEF5', '#FEE4C4', '#E4F4C4', '#F8D2D2', '#E2D2F0']

# Create a horizontal bar graph for the top 10 countries
fig, ax = plt.subplots(figsize=(6,3))
ax.barh(order_volume.index, order_volume.values, color=colors)
ax.invert_yaxis()  # Invert the y-axis to show countries with highest order volume at the top
ax.set_title('Order Volume by Country (Top 10)')
ax.set_xlabel('Number of Orders')
ax.set_ylabel('Country')

plt.show()

pd.DataFrame(order_volume).round(2).head(10)
```

#### Top 10 Non-European Countries by GDP per Capita

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
GDP = pd.read_csv('World Bank/Countries_and_continents_of_the_world.csv')

# Filter out countries in Europe
GDP = GDP.loc[GDP['Continent'] != 'Europe']

# Sort the dataframe by GDP per capita in descending order
GDP = GDP.sort_values('GDP ($ per capita)', ascending=False)

# Select the top 10 countries by GDP per capita
top_10_GDP = GDP.head(10)

# Sort the top 10 data by GDP per capita in ascending order
top_10_GDP = top_10_GDP.sort_values('GDP ($ per capita)')

# Create a horizontal bar chart of the top 10 non-European countries by GDP per capita
plt.barh(top_10_GDP['Country Name'], top_10_GDP['GDP ($ per capita)'], color=colors)

# Set the x-axis label
plt.xlabel('GDP per capita ($)')

# Set the y-axis label
plt.ylabel('Country')

plt.title('Top 10 Non-European Countries by GDP per Capita')

# Display the chart
plt.show()
```

#### Top 10 Non-European Countries by Local Purchasing Power Index

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
Purchasing_Power = pd.read_csv('World Bank/countries-dataset-2020/Cost of living index by country 2020.csv')

# Filter out the columns of interest
cols = ['Country', 'Cost of Living Index', 'Rent Index', 'Local Purchasing Power Index']
Purchasing_Power_filtered = Purchasing_Power[cols]

# Exclude the European Countries
excluded_countries = ['Switzerland', 'Luxembourg', 'Sweden', 'Denmark', 'Germany', 'Finland', 'Netherlands','United Kingdom','Norway']
Purchasing_Power_filtered = Purchasing_Power_filtered[~Purchasing_Power_filtered['Country'].isin(excluded_countries)]

# Sort the filtered dataframe by Local Purchasing Power Index in descending order
Purchasing_Power_filtered = Purchasing_Power_filtered.sort_values('Local Purchasing Power Index', ascending=False)

colors = ['#FEE4E4', '#FECDD3', '#FDC7DE', '#E2D2F0', '#D3E0F5', '#C9DEF5', '#FEE4C4', '#E4F4C4', '#F8D2D2', '#E2D2F0']


# Create a horizontal bar chart of the top 10 countries by Local Purchasing Power Index
fig, ax = plt.subplots()
ax.barh(Purchasing_Power_filtered['Country'].head(10)[::-1], Purchasing_Power_filtered['Local Purchasing Power Index'].head(10)[::-1], color=colors)

# Set the x-axis label
ax.set_xlabel('Local Purchasing Power Index')

# Set the y-axis label
ax.set_ylabel('Country')


ax.set_title('Top 10 Non-European Countries by Local Purchasing Power Index')

# Set the figure size to 6x3 inches
fig.set_size_inches(6, 3)

# Display the chart
plt.show()

```

#### Total Sales by State in the United States (Top 10)

```python
import pandas as pd
import numpy as np

# Load the CSV file into a pandas DataFrame
Supply_Chain = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='unicode_escape')

# Keep only the necessary columns
Supply_Chain = Supply_Chain[['Order City', 'Order State', 'Order Country', 'order date (DateOrders)', 'Order Profit Per Order', 'Order Item Total']]

# Remove rows where Order City, Order State, or Order Country are blank
Supply_Chain = Supply_Chain.dropna(subset=['Order City', 'Order State', 'Order Country'])

Supply_Chain_us = Supply_Chain[Supply_Chain['Order Country'] == 'Estados Unidos']

# Group the DataFrame by Order State and calculate the total sales (i.e., sum of Order Item Total)
sales_by_state = Supply_Chain_us.groupby('Order State')['Order Item Total'].sum().sort_values(ascending=False)[:10]

# Create a pandas DataFrame with the top 10 states and their sales
top_states_Supply_Chain_us = pd.DataFrame({'Order State': sales_by_state.index, 'Total Sales': sales_by_state.values})
top_states_Supply_Chain_us = top_states_Supply_Chain_us.set_index('Order State')

colors = ['#FEE4E4', '#FECDD3', '#FDC7DE', '#E2D2F0', '#D3E0F5', '#C9DEF5', '#FEE4C4', '#E4F4C4', '#F8D2D2', '#E2D2F0']

# Create a horizontal bar chart with light orange bars
fig, ax = plt.subplots()
ax.barh(sales_by_state.index, sales_by_state.values, color= colors)
ax.invert_yaxis()  # Invert the y-axis to show states with highest sales at the top
ax.set_title('Total Sales by State in the United States (Top 10)')
ax.set_xlabel('Total Sales')
ax.set_ylabel('Order State')
plt.show()

top_states_Supply_Chain_us 
```

### Order Frequency by State in the United States (Top 10)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
Supply_Chain_us = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='unicode_escape')

# Keep only the necessary columns
Supply_Chain_us = Supply_Chain_us[['Order City', 'Order State', 'Order Country', 'order date (DateOrders)', 'Order Profit Per Order', 'Order Item Total']]

# Remove rows where Order City, Order State, or Order Country are blank
Supply_Chain_us = Supply_Chain_us.dropna(subset=['Order City', 'Order State', 'Order Country'])

# Convert order date to datetime format
Supply_Chain_us['order date (DateOrders)'] = pd.to_datetime(Supply_Chain_us['order date (DateOrders)'], format='%m/%d/%Y %H:%M')

# Filter orders from United States
us_orders = Supply_Chain_us[Supply_Chain_us['Order Country'] == 'Estados Unidos']

# Group orders by state and calculate the number of unique order dates
order_frequency = us_orders.groupby('Order State')['order date (DateOrders)'].nunique().sort_values(ascending=False)[:10]

# Create a horizontal bar graph with light orange bars for the top 10 states
fig, ax = plt.subplots()
ax.barh(order_frequency.index, order_frequency.values, color= colors)
ax.invert_yaxis()  # Invert the y-axis to show states with highest frequency at the top
ax.set_title('Order Frequency by State in the United States (Top 10)')
ax.set_xlabel('Number of Orders')
ax.set_ylabel('State')
plt.show()

# Display the results as a dataframe
pd.DataFrame(order_frequency, columns=['Order Frequency']).round(2)
```

## Which country shows the best long-term opportunity for expansion? 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate total revenue and profit for each country
Supply_Chain['Total Revenue'] = Supply_Chain['Order Item Total'] * Supply_Chain['Order Profit Per Order']
country_revenue = Supply_Chain.groupby('Order Country')['Total Revenue'].sum()
country_profit = Supply_Chain.groupby('Order Country')['Order Profit Per Order'].sum()

# Calculate average revenue and profit per order for each country
country_order_count = Supply_Chain.groupby('Order Country')['Order City'].nunique()
country_avg_revenue_per_order = country_revenue / country_order_count
country_avg_profit_per_order = country_profit / country_order_count

# Calculate growth potential based on revenue and profit per order
growth_potential = country_avg_revenue_per_order * country_avg_profit_per_order

# Sort countries by growth potential in descending order
sorted_countries = growth_potential.sort_values(ascending=False)

# Create a DataFrame for the top 10 countries with the most potential for growth
top_countries = pd.DataFrame({
    'Country': sorted_countries.index[:10],
    'Growth Potential': sorted_countries.round(2)[:10]
})

# Set the index of the DataFrame to the country name
top_countries.set_index('Country', inplace=True)

# Sort the DataFrame in descending order
top_countries = top_countries.sort_values(by='Growth Potential', ascending=True)

# Plot a horizontal bar chart
ax = top_countries.plot(kind='barh', legend=None, color= colors)

# Set the axis labels and title
ax.set_xlabel('Growth Potential')
ax.set_ylabel('Country')
ax.set_title('Top 10 Countries with the Most Growth Potential')

# Show the plot
plt.show()
```

#### Population of Singapore over time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas dataframe
country_population = pd.read_csv('World Bank/country_population.csv')

# Filter for Singapore
country_population_filtered = country_population[country_population['Country Name'] == 'Singapore']

# Select the columns of interest and fill missing values with the column median
cols = ['Country Name', '1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016']
country_population_filtered = country_population_filtered[cols].fillna(country_population_filtered.median())

# Convert the columns to rows and create a line chart of population over time
country_population_filtered = country_population_filtered.set_index('Country Name')
country_population_filtered = country_population_filtered.T
country_population_filtered = country_population_filtered.astype(int)

country_population_filtered.plot(kind='line')

# Set the title of the chart
plt.title('Population of Singapore over Time')

# Set the x-axis label
plt.xlabel('Year')

# Set the y-axis label
plt.ylabel('Population')

# Display the chart
plt.show()
```
